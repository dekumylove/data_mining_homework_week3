## 数据缺失的处理

### 将缺失部分剔除

```python
df_dropna = df.dropna()
print(df.shape)
print(df_dropna.shape)
print(f'there are {df.shape[0] - df_dropna.shape[0]} items containing nan')
```

```python
q0_dropna = df_dropna[numeric_columns].quantile(0)
q1_dropna = df_dropna[numeric_columns].quantile(0.25)
q2_dropna = df_dropna[numeric_columns].quantile(0.5)
q3_dropna = df_dropna[numeric_columns].quantile(0.75)
q4_dropna = df_dropna[numeric_columns].quantile(1)
null_count_dropna = df_dropna[numeric_columns].isnull().sum()
print(f'最小值分别为:')
print(q0_dropna)
print(f'第一四分位数分别为:')
print(q1_dropna)
print(f'中位数分别为:')
print(q2_dropna)
print(f'第三四分位数分别为:')
print(q3_dropna)
print(f'最大值分别为:')
print(q4_dropna)
print('缺失值个数分别为')
print(null_count_dropna)
print('与原始数据集相比，中位数、第三四分位数和最大值均有所变化，最小值、第一四分位数没有发生变化')
```

### 用最高频率值填充缺失值

```python
stars_count = df['stars_count'].value_counts()
df['stars_count'].fillna(stars_count.index[0], inplace = True)

forks_count = df['forks_count'].value_counts()
df['forks_count'].fillna(forks_count.index[0], inplace = True)

issues_count = df['issues_count'].value_counts()
df['issues_count'].fillna(issues_count.index[0], inplace = True)

pull_requests_counts = df['pull_requests'].value_counts()
df['pull_requests'].fillna(pull_requests_counts.index[0], inplace = True)

contributors_count = df['contributors'].value_counts()
df['contributors'].fillna(contributors_count.index[0], inplace = True)

q0_fillm = df[numeric_columns].quantile(0)
q1_fillm = df[numeric_columns].quantile(0.25)
q2_fillm = df[numeric_columns].quantile(0.5)
q3_fillm = df[numeric_columns].quantile(0.75)
q4_fillm = df[numeric_columns].quantile(1)
null_count_fillm = df[numeric_columns].isnull().sum()
print(f'最小值分别为:')
print(q0_fillm)
print(f'第一四分位数分别为:')
print(q1_fillm)
print(f'中位数分别为:')
print(q2_fillm)
print(f'第三四分位数分别为:')
print(q3_fillm)
print(f'最大值分别为:')
print(q4_fillm)
print('缺失值个数分别为')
print(null_count_fillm)
print('与原始数据集相比，最小值、第一四分位数、中位数、第三四分位数和最大值均没有发生变化')
```

### 通过属性的相关关系来填补缺失值

```python
df['language'] = df.groupby('stars_count')['language'].transform(lambda x: x.fillna(x.value_counts().index[0] if not x.value_counts().empty else 'Unknown'))
print(f'通过对stars_count进行分组，进而在分组内用最高频率值填充缺失值')
language_count_fillna3 = df['language'].value_counts()
print('language填充后取值的频数为:')
print(language_count_fillna3)
```

### 用最高频率值填充缺失值

```python
df_fillmax = df.copy()

stars_count = df_fillmax['stars_count'].value_counts()
df_fillmax['stars_count'].fillna(stars_count.index[0], inplace = True)

forks_count = df_fillmax['forks_count'].value_counts()
df_fillmax['forks_count'].fillna(forks_count.index[0], inplace = True)

watchers_count = df_fillmax['watchers'].value_counts()
df_fillmax['watchers'].fillna(watchers_count.index[0], inplace = True)

pull_requests_counts = df_fillmax['pull_requests'].value_counts()
df_fillmax['pull_requests'].fillna(pull_requests_counts.index[0], inplace = True)

commit_count = df_fillmax['commit_count'].value_counts()
df_fillmax['commit_count'].fillna(commit_count.index[0], inplace = True)

q0_fillm = df_fillmax[numeric_columns].quantile(0)
q1_fillm = df_fillmax[numeric_columns].quantile(0.25)
q2_fillm = df_fillmax[numeric_columns].quantile(0.5)
q3_fillm = df_fillmax[numeric_columns].quantile(0.75)
q4_fillm = df_fillmax[numeric_columns].quantile(1)
null_count_fillm = df_fillmax[numeric_columns].isnull().sum()
print(f'最小值分别为:')
print(q0_fillm)
print(f'第一四分位数分别为:')
print(q1_fillm)
print(f'中位数分别为:')
print(q2_fillm)
print(f'第三四分位数分别为:')
print(q3_fillm)
print(f'最大值分别为:')
print(q4_fillm)
print('缺失值个数分别为')
print(null_count_fillm)
print('与原始数据集相比，最小值、第三四分位数和最大值均没有发生变化，第一四分位数、中位数发生了变化')
```

### 通过属性的相关关系填充缺失值

```python
df_fillrelated = df.copy()

df_fillrelated['commit_count'] = df_fillrelated.groupby('pull_requests')['commit_count'].transform(lambda x: x.fillna(x.value_counts().index[0] if not x.value_counts().empty else 'Unknown'))
print(f'通过对pull_requests进行分组，进而在分组内用最高频率值填充缺失值，填充后柱状图如下')

plt.hist(df_fillrelated['commit_count'], color = 'r', alpha = 0.5)
plt.xlabel('commit_count')
plt.ylabel('value')
plt.show()
```
